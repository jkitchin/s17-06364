# ASSIGNED Catalyst particle effectiveness

The first-order, irreversible reaction (A  → B) takes place in a 0.3 cm radius spherical catalyst pellet at T = 450 K. At 0.7 atm partial pressure of A, the pellet’s production rate is −2.125 × 10<sup>−5</sup> mol/(cm<sup>3</sup> s). The effective diffusivity of A in the pellet is D<sub>A</sub> = 0.007 cm<sup>2</sup>/s.

-   Determine the rate constant and effectiveness factor for this pellet. Hint: You can do this by solving non-linear algebraic equations.

-   Determine the production rate at the same temperature in a 0.15 cm radius spherical pellet.

# solution

We know that:

-   $R_p = \eta k C_{As}$
-   $\Phi = \sqrt{\frac{k a^2}{D_e}}$
-   $ \eta = \frac{1}{\Phi}  \left [\frac{1}{\tanh 3 \Phi} - \frac{1}{3 \Phi} \right]$

There are three unknowns: $k, \eta, \Phi$ and three equations. We can solve them with fsolve.

```python
import numpy as np
from scipy.optimize import fsolve


Rap =  -2.125e-5 # mol / cm^3 / s
rp = 0.3  # cm, particle radius
De = 0.007 # effective diffusivity


Pa = 0.7 # atm
Rc = 82.06 # cm^3 atm / mol /K
T = 450 # K

Cas = Pa / Rc / T

a = rp / 3

def objective(X):
    k, phi, eta = X
    z1 = Rap + eta * k * Cas
    z2 = phi - np.sqrt(k * a**2 / De)
    z3 = eta - 1 / phi * (1 / np.tanh (3 * phi) - 1 / (3 * phi))
    return (z1, z2, z3)

k, phi, eta = fsolve(objective, [2, 2, 0.6])
print('The rate constant is {} 1/s'.format(k))
print('The effectiveness factor is {}.'.format(eta))
```

Now, for the new particle, we just calculate a new &Phi; using the rate constant from above and a new particle radius. We use that &Phi; to estimate the new effectiveness factor. Then we compute the effective rate with $R = \eta k C_{As}$.

```python
R2 = 0.15 # cm
a2 = R2 / 3

phi2 = np.sqrt(k * a2**2 / De)
eta2 = 1 / phi2 * (1 / np.tanh (3 * phi2) - 1 / (3 * phi2))

print('The new rate is {} mol / cm^3 / s.'.format(eta2 * k * Cas))
```
