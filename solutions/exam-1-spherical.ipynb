The ammonia decomposition reaction $2NH_3 \rightarrow  N_2 + 3 H_2$ was run in these reactors. In both cases the inlet pressure was 2000 kPa, and a conversion of 0.81 was achieved with an exit pressure of 1980 kPa in the spherical reactor. The same reaction was run in a tubular packed bed reactor, with the same inlet conditions, the same mass of catalyst, and the same kind of catalyst, but conversion was only 0.71, and the exit pressure was 308 kPa.
Explain why the conversion is higher in the spherical reactor. Be as quantitative as you can, and use equations in your explanation. It is not necessary to solve the equations, but it should be clear from the math what you are communicating.

## Solution

In the spherical reactor the exit pressure is significantly higher than in the tubular reactor. That means there is a substantial pressure drop in the tubular reactor which results in a lower conversion. The lower conversion is due to the following factors in the tubular reactor:

1.  The pressure is lower, so the concentrations are lower ($C=P/RT$), and consequently the reaction rates are lower. This leads to lower conversion.
2.  The volumetric flow is higher: $v = v_0 P_0 / P$. This means the reactants spend less time in the reactor, and have less time to react.

The reason why the pressure drop is lower in the spherical reactor is because of the wider cross-sectional area ($A_C$). You can see in this formula that increasing the cross-sectional area reduces the pressure drop factor because it is in the denominator.
$\frac{dP}{dW} = -\frac{\beta_0}{A_c (1 - \phi) \rho_c}\frac{P_0}{P} \frac{T}{T_0} \frac{F_T}{F_{T0}}$
The larger area also reduces the gas velocity, which reduces the pressure drop factor.	This is evident in
$\beta_0 = \frac{G (1 - \phi)}{\rho_0 g_c D_p \phi^3}\left [ \frac{150 (1 - \phi) \mu}{D_p} + 1.75 G \right]$
where the superficial velocity $u = \nu / A_C$ goes down with the cross-sectional area, as well as the superficial mass velocity $G$ which decreases with increasing cross-sectional area ($G = \rho u$).
