Consider the gas-phase reaction $A + B \leftrightharpoons C$. The equilibrium constant at T=298K is $K=8$. It is exothermic with $\Delta H = -10$ kcal/mol. Assuming the heat of reaction is a constant, and that you start with a 2:1 molar ratio of $A$ and $B$ at 1 atm in a constant volume reactor, calculate the equilibrium pressure of $C$  over the temperature range 300-450K.

If the pressure of a component exceeds its vapor pressure, it will condense to form a liquid. The vapor pressure of $C$ is given by the Clausius-Claperyon equation:

$\ln P_C = c - \frac{\Delta H_{vap}}{RT}$

where $c=7.53$ and $\Delta H_{vap} = 6$ kcal / mol.

Is there a temperature range where you expect to see a liquid in the reactor? If so, what is the range, and explain whether the equilibrium composition you predict is correct or not in that range.
