You are studying the reaction $A + B \rightarrow C$. Unfortunately, you don't know what the heat of reaction is. You design an experiment where an equimolar mixture of $A$ and $B$ is fed at 723K, 1 atm and a volumetric flow of 20 L/s to an adiabatic PFR with a volume of 1000 Liters. The rate constant is known to be $k = 10^{8} \exp(-Ea/(R T)$ L/mol/s, with $Ea=29000$ cal/mol and T in K. You observed that the exit temperature was about 834 K and an exit conversion of about 0.18 was achieved. What is the heat of reaction? Comment on whether your answer makes sense to you.

-   CpA = 38  # cal/mol/K
-   CpB = 20  # cal/mol/K
